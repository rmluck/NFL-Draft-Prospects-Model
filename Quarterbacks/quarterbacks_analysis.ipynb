```python
color_mapping = {
    5: "darkgreen",
    4: "green",
    3: "lightgreen",
    2: "yellow",
    1: "gray",
    0: "lightgray"
}

plt.figure(figsize=(8, 6))
for rating, color in color_mapping.items():
    subset = qbs[qbs["PROSPECT RATING"] == rating]
    plt.scatter(subset["Height"], subset["Hand Size"], color=color, label=f"Prospect Rating {rating}")

plt.xlabel("Height")
plt.ylabel("Hand Size")
plt.title("QBs")
plt.grid(True)
plt.show()
```

```python
qbs = pd.read_csv("quarterbacks.csv")
qbs = qbs[qbs["Draft Class"].isin([2023, 2022, 2021, 2020])]

plt.figure(figsize=(24, 12))
sns.scatterplot(data=qbs, x="40-Yard Dash", y="pressure_to_sack_rate", s=20)
for i in range(len(qbs)):
    plt.text(qbs["40-Yard Dash"][i], qbs["pressure_to_sack_rate"][i], qbs["Player"][i], fontsize=10)

plt.xlabel("40-Yard Dash")
plt.ylabel("Pressure to Sack Rate")
plt.title("QBs")
plt.grid(True)
plt.show()
```
