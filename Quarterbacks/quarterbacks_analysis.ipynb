```python
color_mapping = {
    5: "darkgreen",
    4: "green",
    3: "lightgreen",
    2: "yellow",
    1: "gray",
    0: "lightgray"
}

plt.figure(figsize=(8, 6))
for rating, color in color_mapping.items():
    subset = qbs[qbs["PROSPECT RATING"] == rating]
    plt.scatter(subset["Height"], subset["Hand Size"], color=color, label=f"Prospect Rating {rating}")

plt.xlabel("Height")
plt.ylabel("Hand Size")
plt.title("QBs")
plt.grid(True)
plt.show()
```

```python
def load_image(image_path):
    img = plt.imread(image_path)

    return OffsetImage(img, zoom=0.20)

qbs = pd.read_csv("quarterbacks.csv")
qbs = qbs[qbs["Drafted Round"].isin([1, 2, 3])]
qbs = qbs[qbs["Draft Class"].isin([2023, 2022, 2021, 2020, 2019, 2018])]
qbs["scramble_rate"] = qbs["scrambles_x"] / qbs["dropbacks_x"]

plt.figure(figsize=(24, 12))
ax = sns.scatterplot(data=qbs, x="scramble_rate", y="pressure_to_sack_rate", s=0)

annotations = []
for _, qb in enumerate(qbs):
    annotations.append(ax.text(qb["scramble_rate"] + 0.002, qb["pressure_to_sack_rate"], qb["Player"], fontsize=15))
    imagebox = load_image(os.path.abspath(os.path.join(os.getcwd(), os.pardir)) + "/Logos/" + qb["College"] + ".png")
    ab = AnnotationBbox(imagebox, (qb["scramble_rate"], qb["pressure_to_sack_rate"]), frameon=False)
    ax.add_artist(ab)

adjust_text(annotations)
plt.gca().invert_yaxis()
plt.xlabel("Scramble Rate", fontsize=20)
plt.ylabel("Pressure to Sack Rate", fontsize=20)
plt.title("QBs", fontsize=25)
plt.grid(True)
plt.show()
```
